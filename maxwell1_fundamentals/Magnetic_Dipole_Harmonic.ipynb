# Harmonic Magnetic Dipole

The source code is in [SimPEG Examples](http://docs.simpeg.xyz/en/latest/examples/EM_FDEM_Analytic_MagDipoleWholespace.html)

```python
from SimPEG import *
from SimPEG.Examples import EM_FDEM_Analytic_MagDipoleWholespace as MagDipole

from ipywidgets import interactive, FloatText, FloatSlider, ToggleButtons #interactive plots!

%matplotlib inline
```

## This is the figure in em.geosci.xyz

```python
fig, ax = MagDipole.run()
```

# Change the orientation of the dipole!

```python
Mag_Dipole_Fct = lambda orientation,x,z: MagDipole.run(orientation = orientation, loc = np.r_[x,0.,z])

interactive(Mag_Dipole_Fct, orientation=ToggleButtons(options=['X','Z'],value='Z') , 
                                                       x=FloatText(min=-100.,max=100,step=10.,value=0.), 
                                                       z=FloatText(min=-100.,max=100,step=10.,value=0.))
```
